```python
s = "hello .w@#orld!"

for p in [",", ".", "!", "?", ";", ":"]:
    s = s.replace(p, f" {p} ")
''.join(c for c in s if c.isalnum() or c.isspace() or c in ".,?")
```

```python
news = ''
for c in s:
    if c.isalnum() or c.isspace():
        news += c
    elif c in ".,?":
        news += ' ' + c + ' '

news
```
